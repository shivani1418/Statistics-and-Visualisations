#### To see outliers clearly ,Boxplot 

```python
sns.boxplot(df['residual sugar'])
```

## Z Score Method (to find or remove outlier)

```python
upper_limit= df['residual sugar'].mean()+3*df['residual sugar'].std()
lower_limit= df['residual sugar'].mean()-3*df['residual sugar'].std()
```

```python
print('upper limit:',upper_limit)
print('lower limit:',lower_limit)
```

```python
df.loc[(df['residual sugar']>upper_limit) | (df['residual sugar']<lower_limit)]
```

```python
new_df=df.loc[(df['residual sugar']<upper_limit) & (df['residual sugar']>lower_limit)]
new_df
```

```python
print('lenght of old data:',len(df))
print('lenght of new data:',len(new_df))
print('Outliers:', len(df)-len(new_df))
```

```python
sns.boxplot(new_df['residual sugar'])
```

#### Capping - changing the outliers values to upper or lower limits 

```python
new_df=df.copy()
new_df.loc[(new_df['residual sugar']>upper_limit)]=upper_limit
new_df.loc[(new_df['residual sugar']<lower_limit)]=lower_limit
```

```python
sns.boxplot(new_df['residual sugar'])
```

# IQR method 

```python
q1=df['residual sugar'].quantile(0.25)
q3=df['residual sugar'].quantile(0.75)
iqr=q3-q1
```

```python
print(q1,q3,iqr)
```

```python
upper_limit=q3+1.5*iqr
lower_limit=q1-1.5*iqr
upper_limit,lower_limit
```

```python
sns.boxplot(df['residual sugar'])
```

```python
new_df=df.loc[(df['residual sugar']<upper_limit) & (df['residual sugar']>lower_limit)]
new_df
```

```python
sns.boxplot(new_df['residual sugar'])
```

```python
print('lenght of old data:',len(df))
print('lenght of new data:',len(new_df))
print('Outliers:', len(df)-len(new_df))
```
